```python
from itertools import combinations
from collections import deque
import math
```

```python
def is_prime(number):
    if number == 1:
        return False
    for i in range(2,int(math.sqrt(number))+1):
#         print(number,i)
        if number%i == 0:
            return False
    return number
```

```python
prime_list = []
for n in range(1000,9999):
    prime = is_prime(n)
    if prime:
        prime_list.append(list(str(prime)))
    
```

```python
prime_dict = {}
found_prime = []

for prime in prime_list:
    print(prime)
    
    prime_string = ''.join(prime)
    prime_list_similar = []
    sortedp = sorted(prime)
    
    if sortedp not in found_prime:

        for check in prime_list:
#             print(check)

            if (sortedp == sorted(check)):
                check_int = int(''.join(check))
                prime_list_similar.append(check_int)

        if len(prime_list_similar) >= 3:
            found_prime.append(sortedp)
            prime_dict[prime_string] = prime_list_similar
        
```

```python
def check_list_diffs(L1):
#     list_of_diffs = []
    for first in L1:
        list_of_diffs = []
        for second in L1:
            if first != second:
                diff = abs(first - second)
                if diff not in list_of_diffs:
                    list_of_diffs.append(diff)
                else:
                    return L1
    return False
                
    
        
    
```

```python
answers = []
for prime in prime_dict:
    answer = check_list_diffs(prime_dict[prime])
    if answer:
        answers.append(answer)
    
```
