Problem 10
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
import math
```

```python
def find_prime(number):
    if type(number) != int:
        print('Not a candidate for a prime')
    
    elif number ==1:
        return 0
        
    elif (number == 2 or number == 3):
        return number
    
    factor_list = list(range(2,math.ceil(math.sqrt(number)+1)))
    
    prime = True
    for factor in factor_list:
        if number % factor == 0:
            prime = False
        #return 0 
            
    if prime == True:
        return(number)
            
```

```python
def find_primes_in_range(number):
    
    for candidate in range(number):

        if candidate in (0,1):
            yield 0

        elif (candidate == 2 or candidate == 3):
            yield candidate

        factor_list = list(range(2,math.ceil(math.sqrt(number)+1)))

        prime = True
        for factor in factor_list:
            if candidate % factor == 0:
                prime = False

        if prime == True:
            yield candidate
```

```python
list_of_primes = []
for i in range(2000000):
    prime = find_prime(i)
    if prime:
        if prime >= 2:
            list_of_primes.append(prime)
    else:
        prime = 0
```
