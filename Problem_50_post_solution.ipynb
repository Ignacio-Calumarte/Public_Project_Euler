```python
import math
import numpy as np
```

```python
def is_prime(number):
    if number < 2:
        return None
    for i in range(2,int(math.sqrt(number)+1)):
        if number % i == 0:
            return None
    return number
```

```python
is_prime(30)
```

```python
list_of_primes = []
for i in range(500000):
    prime_number = is_prime(i)
    if prime_number:
        list_of_primes.append(prime_number)
    
```
