```python
import math
```

```python
def find_prime(number):
    if number == 1:
        return False
    for i in range(2,int(math.sqrt(number))+1):
#         print(number,i)
        if number%i == 0:
            return False
    return number
```

```python
def valid_prime(number):
    string = str(number)
    length = len(string)
    
    for n in range(1,length):
        
        new_number = int(string[n:])
        
        if find_prime(new_number) == False:
            return False
        
        new_number_2 = int(string[:-n])
        
        if find_prime(new_number_2) == False:
            return False
    return number
            
    
```

```python
count = 0
n = 11
prime_list = []

while count <11:
    prime = find_prime(n)
    if prime:
        find = valid_prime(prime)
        if find:
            prime_list.append(find)
            count+=1
            print(count)
    n+=2
        
    
```
