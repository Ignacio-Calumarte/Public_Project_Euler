```python
from IPython.core.display import display, HTML
import math
```

```python
display(HTML('<p><i>n</i>! means <i>n</i> × (<i>n</i> − 1) × ... × 3 × 2 × 1</p><p>For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,<br />and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.</p><p>Find the sum of the digits in the number 100!</p>'))
```

```python
number = str(math.factorial(100))
```

```python
def split(word):
    return [int(char) for char in word]
```

```python
list_of_digits = split(number)
```

```python
sum(list_of_digits)
```
