```python
def triangular(number):
    elements = [i for i in range(1,number+1)]
    return sum(elements)
```

```python
def find_factors(number):
    divisors = 0 
    for divisor in range(1,int(number/2)+1):
        if number%divisor == 0:
            divisors+= 1
    return divisors
            
```

```python
triangular(3)
```

```python
find_factors(900000000)
```

```python
def find_factors_efficient(number):
    divisors = 0 
    for divisor in range(1,int(math.sqrt(number))):
        if number%divisor == 0:
            divisors+= 2
    return divisors
```

```python
factors = 0
i=2
j=1
while factors < 500:
    factors = find_factors_efficient(j)
    print(factors,j)
    j+=i
    i+=1
print(i)
```
