```python
maximum = 0
max_perimeter = 0
for p in range(120,1001):
    print(p)
    current_count = 0
    for a in range(1,int(p/2)+1):
        for b in range(1,p-a):
            c = p-(a+b)
            if a**2 + b**2 == c**2:
                print(a,b,c)
                current_count += 1
    print('complete',current_count,maximum,max_perimeter)
    if current_count > maximum:
        maximum = current_count
        max_perimeter = p
max_perimeter
        
                
```
