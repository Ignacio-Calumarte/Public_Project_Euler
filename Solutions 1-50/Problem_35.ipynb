```python
import math
```

```python
def find_prime(candidate):
    if candidate == 1:
        return 0
    
    if candidate == 2:
        return 1
    
    if candidate % 2 == 0:
        return 0
    
    for divisor in range(3,int(math.ceil(math.sqrt(candidate)))+1):
            if candidate%divisor == 0:
                return 0
    return 1
```

```python
def iterate_number(string_number):
#     string = str(integer)
#     for i in range(len(string))
    return string_number[-1] + string_number[:-1]
```

```python
def find_cycle(number):
    if find_prime(number) == 1:
        string = str(number)
        for i in range(len(string)):
#            print(type(string), string)
            string = iterate_number(string)
            if find_prime(int(string)) == 0:
                return False
            else:
                pass
        return number
    else:
        return False
        
    
    
```

```python
list_of_cycles = []
for i in range(2,1000000):
    if find_cycle(i):
        list_of_cycles.append(find_cycle(i))
```
