```python
%%html
<p>145 is a curious number, as 1! + 4! + 5! = 1 + 24 + 120 = 145.</p><p>Find the sum of all numbers which are equal to the sum of the factorial of their digits.</p><p class="smaller">Note: As 1! = 1 and 2! = 2 are not sums they are not included.</p>
```

```python
import math
```

```python
max_sum = 0
for i in range(10):
    max_sum += math.factorial(i)
max_sum
```

```python
sum_numbers = 0
number_list = []

for i in range(3,2000000):
    fact_digits = [math.factorial(int(digit)) for digit in str(i)]
    if sum(fact_digits) == i:
        sum_numbers += i
        number_list.append(i)
        
sum_numbers, number_list
        
```
