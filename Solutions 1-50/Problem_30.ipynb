```python
%%html
<p>Surprisingly there are only three numbers that can be written as the sum of fourth powers of their digits:</p><blockquote>1634 = 1<sup>4</sup> + 6<sup>4</sup> + 3<sup>4</sup> + 4<sup>4</sup><br />8208 = 8<sup>4</sup> + 2<sup>4</sup> + 0<sup>4</sup> + 8<sup>4</sup><br />9474 = 9<sup>4</sup> + 4<sup>4</sup> + 7<sup>4</sup> + 4<sup>4</sup></blockquote><p class="smaller">As 1 = 1<sup>4</sup> is not a sum it is not included.</p><p>The sum of these numbers is 1634 + 8208 + 9474 = 19316.</p><p>Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.</p>
```

```python
sum_of_powers = 0
sum_of_powers_list = []
for i in range(2,300000):
    digit_list = [int(digit)**5 for digit in str(i)]
    temp_sum_of_powers = sum(digit_list)
    if temp_sum_of_powers == i:
        sum_of_powers+=i
        sum_of_powers_list.append(i)
sum_of_powers
        
```
