```python
import calendar
from datetime import date
import pandas as pd

# get today's date
d = date.today()
print('Date is:', d)

# get day name in english
x = calendar.day_name[d.weekday()]
print('Weekday name is:', x)
```

```python
def day_of_week(date):
    return calendar.day_name[pd.Timestamp(date).weekday()]
```

```python
day_of_week('1900-1-3')
```

```python
day_of_week('2202-7-12')
```

```python
list_of_dates = []
for year in range(1901,2001):
    for month in range(1,13):
        date = str(year)+'-'+str(month)+'-01'
        list_of_dates.append(date)
        
```

```python
count = 0
for date in list_of_dates:
    if day_of_week(date) == 'Saturday':
        count+=1
count
    
```
