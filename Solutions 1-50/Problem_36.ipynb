```python
number_list = []
number_sum = 0

for number in range(10**6):
    binary_number_string = bin(number)[2:]
    if binary_number_string[0] != '0':
    
        if binary_number_string == binary_number_string[::-1]:
            base10_string = str(number)

            if base10_string == base10_string[::-1]:
                number_list.append(number)
                number_sum += number
                
number_sum, number_list
        
```
