Question 5:
2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
primes = [2,3,5,7,11,13,17,19]
```

```python
for i in range(20,1,-1):
    if i not in primes:
        for divisor in primes:
            if i % divisor == 0:
                remainder = int(i / divisor)
                if remainder not in primes:
                    
                    print(i, divisor, remainder)
```

```python
def find_factors(number,list_of_factors):
    factors = []
    if number not in list_of_factors:
        for divisor in list_of_factors:
            if number % divisor == 0:
                remainder = int(number / divisor)
                factors.extend([divisor,remainder])
    factors_2 = factors.copy()
    
    for factor in factors_2:
        if factor not in list_of_factors:
            factors_2.remove(factor)
            for divisor in list_of_factors:
                if factor % divisor == 0:
                    remainder = int(factor / divisor)
                    factors_2.extend([divisor,remainder])
            
    return factors,factors_2
```

```python
find_factors(20,primes)
```
