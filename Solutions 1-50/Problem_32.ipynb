```python
%%html
<p>We shall say that an <var>n</var>-digit number is pandigital if it makes use of all the digits 1 to <var>n</var> exactly once; for example, the 5-digit number, 15234, is 1 through 5 pandigital.</p>

<p>The product 7254 is unusual, as the identity, 39 × 186 = 7254, containing multiplicand, multiplier, and product is 1 through 9 pandigital.</p>

<p>Find the sum of all products whose multiplicand/multiplier/product identity can be written as a 1 through 9 pandigital.</p>

<div class="note">HINT: Some products can be obtained in more than one way so be sure to only include it once in your sum.</div>

```

```python
import random
import numpy as np
```

```python

# 2 digit number    
list_of_digits = ['1','2','3','4','5','6','7','8','9']
list_of_results = []

for first in list_of_digits:
#     list_of_digits = [1,2,3,4,5,6,7,8,9]
    second_list = list_of_digits.copy()
    second_list.remove(first)
#     print('secpnd',second_list)
    
    for second in second_list:
        third_list = second_list.copy()
        third_list.remove(second)
#         print('third',third_list)
        
#         print(first,second,new_list,third_list)
        
# 3 digit number
        for third in third_list:
            fourth_list = third_list.copy()
            fourth_list.remove(third)
#             print('fourth',fourth_list)
        
                
            for fourth in fourth_list:
                fifth_list = fourth_list.copy()
                fifth_list.remove(fourth)
#                 print('fifth',fifth_list)
                
                for fifth in fifth_list:
                    last_list = fifth_list.copy()
                    last_list.remove(fifth)
#                     print((first+second),(third+fourth+fifth),last_list)
                    
                    multiplicand = int(first+second)
                    multiplier = int(third+fourth+fifth)
                    product = multiplicand*multiplier
                    product_target = set(last_list)
                    
                    if len(str(product)) == len(last_list):
                        if set(str(product)) == set(last_list):
                            if product not in list_of_results:
                                list_of_results.append(product)
                                
                    
                    multiplicand = int(first)
                    multiplier = int(second+third+fourth+fifth)
                    product = multiplicand*multiplier
                    product_target = set(last_list)
                    
                    if len(str(product)) == len(last_list):
                        if set(str(product)) == set(last_list):
                            if product not in list_of_results:
                                list_of_results.append(product)
                                
                    multiplicand = int(first)
                    multiplier = int(second+third+fourth)
                    product = multiplicand*multiplier
                    product_target = set(fifth_list)
                    
                    if len(str(product)) == len(fifth_list):
                        if set(str(product)) == set(fifth_list):
                            if product not in list_of_results:
                                list_of_results.append(product)
                    
```
