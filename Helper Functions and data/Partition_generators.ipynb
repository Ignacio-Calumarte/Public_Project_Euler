```python
q = { 1: [[1]] }

def decompose(n):
    try:
        return q[n]
    except:
        pass

    result = [[n]]

    for i in range(1, n):
        a = n-i
        R = decompose(i)
        for r in R:
            if r[0] <= a:
                result.append([a] + r)

    q[n] = result
    return result
```

```python
def partitions(n):
    parts = [1]+[0]*n
    print(parts)
    for t in range(1, n+1):
        for i, x in enumerate(range(t, n+1)):
            print(parts[x],parts[i])
            parts[x] += parts[i]
    return parts[n]
```
