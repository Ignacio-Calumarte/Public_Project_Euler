```python
def fib_gen(N):
    
    first = 0
    second = 1
    count = 0
    
    while count < N:
        
        if count == 0:
            yield(0)
            count += 1
            
        if count in (1,2):
            yield(1)
            count += 1
            
        fib_num = first + second
        
        yield(fib_num)
        
        first = second
        second = fib_num
        count += 1
        
        
```

```python
max(list(fib_gen(5000)))
```
