```python
import math
```

```python
def find_prime(number):
    if number == 1:
        return False
    for i in range(2,int(math.sqrt(number))+1):
#         print(number,i)
        if number%i == 0:
            return False
    return number
```

```python
for i in range(100):
    y = find_prime(i)
    if y:
        print(y)
```
