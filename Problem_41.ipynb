```python
import math
import itertools
```

```python
import itertools
inp_list = [4, 5, 6]
permutations = list(itertools.permutations(inp_list))
```

```python
def is_prime(number):
    for i in range(2,int(math.sqrt(number)+1)):
        if number % i == 0:
            return False
    return True
```

```python
is_prime(10)
```

```python
for i in range(9,5,-1):
    inp_list = ['9','8','7','6','5','4','3','2','1']
#     inp_list[::-1][:i][::-1]
#     print(inp_list)
    permutations = list(itertools.permutations(inp_list[::-1][:i][::-1]))
    for permutation in permutations:
        number = int(''.join(permutation))
        if is_prime(number):
            print(number)
            break
```
