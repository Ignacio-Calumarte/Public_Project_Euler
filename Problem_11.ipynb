```python
for i,row in enumerate(grid):
    grid[i] = row.split()
    int_row = map(int,grid[i])
    grid[i] = list(int_row)
```

```python
q11_array = np.array(grid)
```

```python
def diagonal_down(array,length):
    total = 0
    dimensions = array.shape
    
    for row in range(dimensions[0]-(length-1)):
        for column in range(dimensions[1]-(length-1)):
            cells = []
            for i in range(length):
                cell = array[row+i][column+i]
                cells.append(cell)
            temp_total=(np.array(cells)).prod() 
            
            if temp_total > total:
                total = temp_total
                
    return total
            
    
                
```

```python
diagonal_down(q11_array,4)
```

```python
def diagonal_up(array,length):
    total = 0
    dimensions = array.shape
    length = length -1
    
    for row in range(length,dimensions[0]):
        for column in range(dimensions[1]-length):
            cells = []
            for i in range(length+1):
                cell = array[row-i][column+i]
                cells.append(cell)
            
            temp_total=(np.array(cells)).prod() 
            if temp_total > total:
                total = temp_total
                
    return total
```

```python
diagonal_up(q11_array,4)
```

```python
def generate_elements(array):
    for x in range(array.shape[0]):
        for y in range(array.shape[1]):
            yield(array[x][y])
```

```python
def horozontal_sum(array, length):
    total= 0 
    dimensions = array.shape
    
    for row in range(dimensions[0]):
        for column in range(dimensions[1]-(length-1)):
            temp_total = (array[row][column:column+(length)]).prod()
#            print(row,column,column+length)
            if temp_total > total:
                total = temp_total
    return total
            
            
        
    
    
```

```python
horozontal_sum(q11_array,4)
```

```python
def vertical_elements(array, length):
    total= 0 
    dimensions = array.shape
    
    for row in range(dimensions[0]-(length-1)):
        for column in range(dimensions[1]):
            temp_total = array[row:row+4,column].prod()
#             print(row,row+4,column)
            if temp_total > total:
                total = temp_total
    return total
```
